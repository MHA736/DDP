```python
#Membuat diagram Venn

import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Definisikan himpunan
himpunan_A = {1, 2, 3, 4, 5}
himpunan_B = {3, 4, 5, 6, 7}

komplemen_A = himpunan_A-himpunan_B
irisan = himpunan_A & himpunan_B
komplemen_B = himpunan_B-himpunan_A

# Buat diagram Venn
venn = venn2([himpunan_A, himpunan_B], set_labels=('A', 'B'))
Venn.get_label_by_id('10').set_text(f"A-B={komplemen_A}")
venn.get_label_by_id('01').set_text(f"B-A={komplemen_B}")
venn.get_label_by_id('11').set_text(f"A n B={irisan}")

plt.show()
```

```python
# Mendefinisikan fungsi untuk mencari akar persamaan kuadrat
import cmath
def akar_persamaan_kuadrat(a, b, c):
  delta = (b**2) - 4*(a*c)
  akar1 = (-b - cmath.sqrt(delta)) / (2*a)
  akar2 = (-b + cmath.sqrt(delta)) / (2*a)
  return akar1, akar2

# Mencari akar dari persamaan p^2 - 3p - 810 = 0
a = 1
b = -3
c = -810
akar1, akar2 = akar_persamaan_kuadrat(a, b, c)

# Menampilkan hasil
print("Akar persamaan kuadrat adalah:", akar1, "dan", akar2)

# Mencari panjang dan lebar
panjang = akar1.real  # Mengambil nilai real dari akar1
lebar = panjang - 3

# Menampilkan hasil
print("Panjang lahan:", panjang, "meter")
print("Lebar lahan:", lebar, "meter")
```
